```python
# for teng_model

import matplotlib.pyplot as plt

total_counts =  np.array([20]*39)
correct_counts = total_counts - wrong_counts_teng

accuracies_teng = [correct / total for correct, total in zip(correct_counts, total_counts)]

gesture_labels = [str(i) for i in range(1, 40)]

plt.figure(figsize=(10, 6))
plt.plot(gesture_labels, accuracies_teng, marker='o', linestyle='-', color='b')
plt.title('Test Accuracy for Different Gestures teng_model')
plt.xlabel('Gesture')
plt.ylabel('Accuracy')
plt.xticks(gesture_labels)
plt.ylim(0, 1)
plt.grid(True)

plt.show()
```
