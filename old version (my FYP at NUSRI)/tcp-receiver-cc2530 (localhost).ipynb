```python
'''
给板子烧录的arduino程序,只负责简单的读取
如果想实现采集一段时间内的平均值，在python里实现
'''
import serial
import time
import numpy as np

# 防止出现数据突然过大或者突然过小的情况
def normal(teng_data):
    for i in teng_data:
        if float(i)<10 or float(i)>25:
            return False
    return True

'''
以下,计算5秒内采样到的上百组数据平均值
'''
data = []

ser=serial.Serial("COM4",115200,timeout=1)
start=time.time()

num=0 
while 1:
    value=ser.readline().decode("utf-8")
    value=value.split(" ") # 让每个数据呈现出来
    teng_data=value[0:5]
    if value[0]!='' and len(value)==11 and normal(teng_data)==True and (float(value[8])>=0.01 or float(value[8])<=-0.01): 
        print(value)
        data.append(value)
    end=time.time()
    
    if end-start>5:
        break
```

```python
data = np.array(data)
for i in range(len(data)):
    data[i][10] = data[i][10].strip('/r/n')
    for j in range(len(data[0])):
        data[i][j] = float(data[i][j])
data=data.astype("float32")

mean = np.array([np.mean(data[:,0]),np.mean(data[:,1]),np.mean(data[:,2]),np.mean(data[:,3]),np.mean(data[:,4]),
                 np.mean(data[:,5]),np.mean(data[:,6]),np.mean(data[:,7]),np.mean(data[:,8]),np.mean(data[:,9]),
                 np.mean(data[:,10])])
```
