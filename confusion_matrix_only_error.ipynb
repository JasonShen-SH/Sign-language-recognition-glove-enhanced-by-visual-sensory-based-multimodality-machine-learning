```python
# for teng_model

import matplotlib.pyplot as plt

misclassification_matrix = [[0] * 39 for _ in range(39)]

for true_class in range(0,39,1):
    misclassified_samples = error_teng[true_class]
    if misclassified_samples!=[]:
        for i in range(0,len(misclassified_samples),1):
            tmp = misclassified_samples[i]
            misclassification_matrix[true_class - 1][tmp - 1] += 1

x, y = [], []
for true_class in range(1, 40):
    for misclassified_class in range(1, 40):
        count = misclassification_matrix[true_class - 1][misclassified_class - 1]
        if count > 0:
            x.append(true_class)
            y.append(misclassified_class)
            
x_teng = x
y_teng = y

plt.scatter(x_teng, y_teng, marker='o', color='red', alpha=0.5)
plt.xlabel('True Class')
plt.ylabel('Misclassified Class')
plt.title('Misclassified Samples teng_model')
plt.xticks(range(1, 40))
plt.yticks(range(1, 40))
plt.grid(True)
plt.show()
```

```python
# for teng_imu_model

error_teng_imu = np.array([[13,24],[25,28],[6,8],[4,19]])

misclassification_matrix = [[0] * 39 for _ in range(39)]

for i in range(error_teng_imu.shape[0]):
    row = error_teng_imu[i][0]
    col = error_teng_imu[i][1]
    misclassification_matrix[row - 1][col - 1] += 1

    
x, y = [], []
for true_class in range(1, 40):
    for misclassified_class in range(1, 40):
        count = misclassification_matrix[true_class - 1][misclassified_class - 1]
        if count > 0:
            x.append(true_class)
            y.append(misclassified_class)
            
x_teng_imu = x
y_teng_imu = y

plt.scatter(x_teng_imu, y_teng_imu, marker='o', color='red', alpha=0.5)
plt.xlabel('True Class')
plt.ylabel('Misclassified Class')
plt.title('Misclassified Samples teng_imu_model')
plt.xticks(range(1, 40))
plt.yticks(range(1, 40))
plt.grid(True)
plt.show()
```
